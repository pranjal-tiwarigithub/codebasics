### scaling each pixel value in array [0,1]

```python
x_train_flatten = x_train.reshape(len(x_train),28*28) /255  # value of x between 0 to 255 
x_test_flatten = x_test.reshape(len(x_test),28*28) / 255 
```

```python
model = keras.Sequential([
    keras.layers.Dense(10 , input_shape=(784,), activation = 'sigmoid') # 10 output layer , input_shape , activation func        ... dense layer means every node of this node  connected to next node 
])


model.compile(
    optimizer='adam', # helps to reach golab maximum during back propogation effiecently
    loss='sparse_categorical_crossentropy',
    metrics = ['accuracy'])


model.fit(x_train_flatten, y_train , epochs = 5)
```

```python
model.evaluate(x_test_flatten , y_test)
```

```python
y_predicted = model.predict(x_test_flatten)
```
