```python
def log_loss(y_predicted , y_true):
    y_predicted_new = [max(epsilon ,i) for i in y_predicted] ## remove all 0
    y_predicted_new = [min(1-epsilon , i ) for i in y_predicted_new ] ## remove all 1 
    y_predicted_new = np.array(y_predicted_new)
    return - np.mean(y_true * np.log(y_predicted_new) + ((1-y_true) * np.log(1- y_predicted_new) ))
    # wrong return - np.mean(np.dot(y_true , np.log(y_predicted_new)) + ( np.dot((1-y_true) , np.log(1- y_predicted_new) )))
    
    
```

```python
log_loss(y_predicted , y_true)
```

## MSE

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python


def MAE_own(y_predicted , y_true ):
    
    error_sum = 0 
    
    for y , y_pred in zip(y_true,y_predicted):
        error_sum += (y - y_pred ) ** 2
        
    print(error_sum)
    return error_sum / len(y_predicted)
    
    
```

```python
MAE_own(y_predicted , y_true )
```
