```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
# create dummt variable columns 


dumies = pd.get_dummies(df.town)

dumies
```

```python
# concate this datafram t5ooriginal df

merged = pd.concat([df,dumies],axis='columns')
merged
```

```python
# # drop  towncolumns 
# drop one of this columns 
# dummy variable trap -  whenever one of variable can be dserived from rest of varaibles , these variables are called
# multicolinear and when we have multi colinearltiy in dataset it canmess upour mnodel
# hence rule is to drop 1 dummy variable
#

final = merged.drop(['town','west windsor'],axis='columns')
```

```python
final
```

```python
# linear regression drops i column automatically , its awre of dummy variable trap

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
x = final.drop(['price'],axis = 'columns') #area , monore , robinsilve
x
```

```python
y = final['price'] # or y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]]) # 0,1 means robinsilve
```

```python
model.predict([[3400,0,0]]) # west windsor
```

```python
model.score(x,y)
```

```python
# modelis 0.95 % correct
```
