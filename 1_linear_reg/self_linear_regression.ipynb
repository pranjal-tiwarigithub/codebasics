```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
# plot a scatter plot

%matplotlib inline

plt.scatter(df.area,df.price , color = 'red', marker = '*')
plt.xlabel('area(sqr ft)')
plt.ylabel('price ($US)')
```

distribution suitable for linear regression 

```python
reg =linear_model.LinearRegression()
reg.fit(df[['area']],df.price) # x takes 2d array 
```

```python
reg.predict([[3300]])
```
